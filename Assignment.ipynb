#### Histogram

```python
sns.set_style("whitegrid")
```

```python
plt.hist(df.Quantity);
plt.xlabel('Quantity')
plt.ylabel('Number of orders')
plt.title("Histogram showing Quantity wise number of orders")
```

```python
plt.hist(df.Discount);
plt.xlabel('Discount');
plt.ylabel('Number of orders')
plt.title("Histogram showing Discount wise number of orders")
```

#### Box Plot

```python
df['Sales'].plot(kind='box')
plt.title("Box plot for sales")
plt.ylabel("Order Value")
```

#### New column

```python
df['Sale per order'] = df['Sales']/df['Quantity'];
print(df['Sale per order']);
```

```python
df['Profit per item'] = df['Profit']/df['Quantity'];
print(df['Profit per item']);
```

#### Count Plot

```python
sns.countplot(x ='Ship Mode', data = df)
plt.ylabel('Count')
plt.title('Frequency of different Shipping Modes')
```

```python
sns.countplot(x ='Category', data = df)
plt.ylabel('Count')
plt.title('Count of category of order')
```
